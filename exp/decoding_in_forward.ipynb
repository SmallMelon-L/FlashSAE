```python
import os
import sys
sys.path.insert(0, '..') 

import torch
import torch.sparse

from tqdm.notebook import tqdm

from src.sae_fuction import topk_sae, topk_sae_fwd_sparse_coo, topk_sae_fwd_sparse_csr, topk_sae_sparse_fused
from src.utils import compare_reconstruct, compare_reconstruct_grad, compare_latency
from src.config import TestConfig

device='cuda'
dtype=torch.float32
```

```python
d_model = 4096
batch_size = 4096
atol = 1e-6
rtol = 1e-6

k = 128

expansion_factor_list = range(4, 33, 4)
```

```python
time_base = []
time_sparse = []

for expansion_factor in tqdm(expansion_factor_list):
    config = TestConfig(
        batch_size=batch_size,
        d_model=d_model,
        expansion_factor=expansion_factor,
        k=k,
        preheat_repeat=10,
        timing_repeat=10,
    )
    
    compare_reconstruct_grad(
        sae_fwd_1=topk_sae,
        sae_fwd_2=topk_sae_fwd_sparse_csr,
        config=config,
        verbose=False,
        atol=1e-6,
        rtol=1e-6,
    )

    fwd_bwd_times_base, fwd_bwd_times_sparse, times_base, times_sparse = compare_latency(
        sae_fwd_1=topk_sae,
        sae_fwd_2=topk_sae_fwd_sparse_csr,
        config=config,
        verbose=False,
    )
    
    time_base.append(times_base['fwd_decode'])
    time_sparse.append(times_sparse['fwd_feature_acts_to_sparse'] + times_sparse['fwd_sparse_decode'])
```

```python
save_result_path = '/inspire/hdd/global_user/hezhengfu-240208120186/jx_project/ml_sys_course_pj/result/decoding_in_forward'

os.makedirs(save_result_path, exist_ok=True)

torch.save(time_base, os.path.join(save_result_path, 'time_base.pt'))
torch.save(time_sparse, os.path.join(save_result_path, 'time_sparse.pt'))
```
